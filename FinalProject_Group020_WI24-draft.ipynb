```python
def fetch_rental_listings(zip_code,option):
    # Replace `your_target_url` with the URL of the site you have permission to scrape
    url = f"http://zillow.com/san-diego-ca-{zip_code}/rentals"
    #example: www.zillow.com/san-diego-ca-92101/rentals/
    if option == 2:
        url = f"http://zillow.com/san-diego-ca-{zip_code}/sale"
        #example: www.zillow.com/san-diego-ca-92101/sale/
    elif option ==3:
        url = f"http://zillow.com/san-diego-ca-{zip_code}/sold"
        #example: www.zillow.com/san-diego-ca-92101/sold/
    
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Referer': 'http://www.google.com/',
    'DNT': '1',  # Do Not Track Request Header
    'Connection': 'keep-alive'
    }

    response = requests.get(url, headers=headers)
    # Send a GET request to the URL
    #response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    
    # Check if the request was successful
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        listings = []
            
        if option > 1:
            ## all the items is in this class
            property_cards = soup.find_all('div', class_='StyledPropertyCardDataWrapper-c11n-8-100-1__sc-hfbvv9-0')
            for card in property_cards:
                ## extract each info we need
                price_tag = card.find('span', {'data-test': 'property-card-price'})
                price = price_tag.text if price_tag else 'N/A'
                
                # The data for bedrooms, bathrooms, squareft
                details_tag = card.find('ul', {'class': 'StyledPropertyCardHomeDetailsList-c11n-8-100-1__sc-1j0som5-0 kAPEO'})
                details = details_tag.find_all('li') if details_tag else []
                
                bedrooms = details[0].text if len(details) > 0 else 'N/A'
                bathrooms = details[1].text if len(details) > 1 else 'N/A'
                square_feet = details[2].text if len(details) > 2 else 'N/A'
                
                # Store in a dictionary and append it
                property_info = {'ZIP Code': zip_code,'price': price,'bedrooms': bedrooms,'bathrooms': bathrooms,'square_feet': square_feet,}
                listings.append(property_info)
                ##print(property_info)
        else:
            # Find all span tags with the specific data-test attribute
            price_tags = soup.find_all('span', {'data-test': 'property-card-price'})
            # If soldFlag is True, we just return the price information
            for tag in price_tags:
                # Extract the text from each tag, which includes the price and possibly other details like number of beds
                price_info = tag.text
                
                listings.append({'ZIP Code': zip_code, 'Price Info': price_info})
                # print out price info
                #print(f"Price Info: {price_info}")

        return listings
            
    else:
        #print(f"Failed to fetch listings for zipcode: {zip_code} status code: {response.status_code}")
        return {}
```

```python
def get_price(value):
    parts = value.split()
    price = ''.join(filter(str.isdigit, parts[0]))
    return price
```

```python
def get_bedrooms(value):
    parts = value.split()
    bedrooms = np.nan
    if 'bd' in value:
        # Extract the substring before 'bd' and get the last word (number of bedrooms)
        bedrooms = value.split('bd')[0].split()[-1]
    elif 'Studio' in value:
        # Set bedrooms to 0 for studios
        bedrooms = 'Studio'
    return bedrooms
```
